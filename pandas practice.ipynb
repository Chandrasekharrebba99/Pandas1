```python
def findNaN(list1):
    mcount = 0
    fcount = 0
    for i in list1:
        if(i == "male"):
            mcount +=1
        if(i == "female"):
            fcount +=1
    return "mail count is",mcount, "Femail count is",fcount
findNaN(df["Sex"])
```
